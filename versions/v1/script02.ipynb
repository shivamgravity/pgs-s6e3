# XG Boost

This time we are working with a boosting algorithm, right off the bat. It definitely would improve a lot in the predictive performance in comparison to the logistic regression, for sure.

Let's find out!

```python
# import required librarires
import sys
sys.path.append('../..')
import pandas as pd
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from preprocessing.script01 import main as preprocess_main
```

```python
train, test = preprocess_main()
X_train = train.drop(columns=['Churn'])
y_train = train['Churn']
X_test = test.drop(columns=['id'])
test_id = test['id']
```

```python
# Feature scaling
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# XGBoost model
model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    random_state=42,
    eval_metric="logloss"
)

model.fit(X_train, y_train)
```

```python
# Prediction
y_pred = model.predict(X_test)
result = pd.DataFrame({
    'id': test_id, 'Churn': y_pred
    })
result.to_csv('../../submissions/submission02.csv', index=False)
```
